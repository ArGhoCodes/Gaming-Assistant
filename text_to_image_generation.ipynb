```python
import os
import torch
from diffusers import StableDiffusionPipeline

class CFG:
    device = "cuda" if torch.cuda.is_available() else "cpu"
    seed = 42
    generator = torch.Generator(device).manual_seed(seed) if torch.cuda.is_available() else torch.Generator().manual_seed(seed)
    image_gen_steps = 35
    image_gen_model_id = "stabilityai/stable-diffusion-2"
    image_height = 1080  # Set the desired image height
    image_width = 1920   # Set the desired image width

    # Ensure torch_dtype is set based on the device
    if device == "cuda":
        torch_dtype = torch.float16
    else:
        torch_dtype = torch.float32  # Use float32 for CPU

# Initialize the pipeline with appropriate dtype and device
pipeline = StableDiffusionPipeline.from_pretrained(
    CFG.image_gen_model_id,
    torch_dtype=CFG.torch_dtype
)

# Move pipeline to the correct device
pipeline = pipeline.to(CFG.device)

# Example usage
# Generate an image with the pipeline (example prompt)
prompt = input("Enter the prompt to generate the image: ")
generated_images = pipeline(
    prompt,
    num_inference_steps=CFG.image_gen_steps,
    generator=CFG.generator,
    height=CFG.image_height,
    width=CFG.image_width
).images

# Save or display the generated image
generated_image = generated_images[0]
generated_image.save("generated_image.png")

print("Image generated and saved as 'generated_image.png'")
```
